# 查詢
在 DataFrame 中若要取值可以使用常用的 loc、iloc 方法，以下分別介紹兩者使用方式。

loc 語法使用方式：loc(列索引範圍, [欄位名稱])

```python
# 印出 index 2-10 資料的學校代碼、學校名稱和總計欄位
print(df.loc[2:15, ['學校代碼', '學校名稱','縣市名稱', '總計']])
```

```python
# iloc 語法使用方式：iloc(列索引範圍（不含結束）, 欄位索引範圍])
```

```python
# 印出 index 0-2 資料的 0:學校代碼、1: 學校名稱和 4:總計欄位
print(df.iloc[0:3, [0, 1, 4]])
```

# 過濾
在取值時我們有時會希望透過設立條件可以過濾出我們想要看的資料，此時可以使用過濾語法，選定欄位和條件進行過濾（把條件當作參數傳入，DataFrame[條件]）：

```python
# 過濾取出欄位總計 大於 20000 的資料
filtered_df = df[df['總計'] > 20000]

print(filtered_df)
```

```python
#由於資料的欄位過多所以不容易看出資訊，所以我們進一步取出較少欄位，可以看出更明顯的資料內容（先只取：'學校名稱', '日間∕進修別', '等級別', '總計', '男生計', '女生計' 欄位）：

# 取出所有列 index 欄位為 學校名稱', '日間∕進修別', '等級別', '總計', '男生計', '女生計' 資料
print(filtered_df.loc[:, ['學校名稱', '日間∕進修別', '等級別', '總計', '男生計', '女生計']])
```

```python
# 我們可以看到 108 學年度大專院校中以 淡江大學 和 中國文化大學 大學部人數是唯二超過兩萬人的學校，男女比也接近 1:1 的比例，算是蠻平均的，其中文化大學的女生甚至比男生還多。
```

# 群組
若我們希望像是 SQL Group By 語法一樣，針對特定欄位進行群組化，我們可以使用 groupby(['欄位名稱'])，同時可以使用 sum() 彙總函式搭配排序來呈現資料：

```python
print(df.groupby(['學校名稱']).sum().sort_values('總計', ascending=False))
```

# 刪除欄位
若希望刪除特定欄位，可以使用 drop([欄位], axis=指定欄或列) 方法（axis=1 為欄），指定要刪除的欄位。

```python
# 準備傳入 DataFrame 的資料
data_1 = {
    'name': ['王小明', '張小華', '廖丁丁', '許小光'],
    'email': ['min@gmail.com', 'hchang@gmail.com', 'laioding@gmail.com', 'hsulight@gmail.com'],
    'grades': [60, 77, 92, 43]
}


# 建立 DataFrame 物件
student_df_1 = pd.DataFrame(data_1)
print(student_df_1)

# 使用 drop 指定欄位，記得要給定 axis=1 為欄。若 axis=0 為代表列
student_df_1 = student_df_1.drop(['grades'], axis=1)
print(student_df_1)
```

# 充填 NA/NaN 值
有時我們建立或讀取進來的資料值中會有 NA/NaN 值（可能是資料遺漏或是沒有值），此時我們可以使用 DataFrame.fillna 來填充 NA/NaN 值，例如改為 0 等。

使用 Numpy 和 Pandas 建立一個有 NA/NaN 值的 DataFrame：

```python
# 引入 pandas 套件，使用別名 pd 可以少打字
import pandas as pd
import numpy as np

df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, 5],
                   [np.nan, 3, np.nan, 4]],
                  columns=list('ABCD'))

print(df)
```

```python
df.fillna(0)
```

# 合併與更新物件
若我們希望合併不同的 DataFrame 我們可以使用列合併 concat() 或是欄位合併 merge()。

以下先建立學習資料：

```python
# 準備傳入 DataFrame 的資料
data_1 = {
    'name': ['王小明', '張小華', '廖丁丁', '許小光'],
    'email': ['min@gmail.com', 'hchang@gmail.com', 'laioding@gmail.com', 'hsulight@gmail.com'],
    'grades': [60, 77, 92, 43]
}

data_2 = {
    'name': ['黃明明', '汪新新', '鮑呱呱', '江組組'],
    'email': ['ww@gmail.com', 'cc@gmail.com', 'bb@gmail.com', 'ee@gmail.com'],
    'grades': [70, 17, 32, 43]
}

# 建立 DataFrame 物件
student_df_1 = pd.DataFrame(data_1)
student_df_2 = pd.DataFrame(data_2)

print(student_df_1)
print(student_df_2)
```

```python
#使用 concat 列合併資料
student_fg_3 = pd.concat([student_df_1, student_df_2])

print(student_fg_3)
```

```python
# 若有 ignore_index 則列的 index 會重新排列：
student_fg_3 = pd.concat([student_df_1, student_df_2], ignore_index=True)

print(student_fg_3)
```

```python
#建立學習資料

# 準備傳入 DataFrame 的資料
data_1 = {
    'name': ['王小明', '張小華', '廖丁丁', '許小光'],
    'email': ['min@gmail.com', 'hchang@gmail.com', 'laioding@gmail.com', 'hsulight@gmail.com'],
    'grades': [60, 77, 92, 43]
}

data_2 = {
    'name': ['王小明', '張小華', '廖丁丁', '許小光'],
    'age': [19, 20, 32, 43]
}

# 建立 DataFrame 物件
student_df_1 = pd.DataFrame(data_1)
student_df_2 = pd.DataFrame(data_2)

print(student_df_1)
print(student_df_2)
```

```python
#使用 merge 合併欄位資料
#有點類似 SQL JOIN 的感覺，根據相同的欄位合併資料：

student_fg_3 = pd.merge(student_df_1, student_df_2)

print(student_fg_3)
```
